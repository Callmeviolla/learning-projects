```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
infl = pd.read_excel('Infl_Thai.xlsx')
infl
```

```python
infl.info()
```

```python
periods = pd.period_range(start='1966-01-01', end='2022-03-31', freq='M')
periods
```

```python
infl.index = periods
del infl['Date']
plt.figure(figsize=(12,6))
infl['Infl'].plot()
plt.show()
```

```python
infl.describe()
```

```python
infl_pred_naiv = infl.shift(1)
infl_pred_naiv
```

```python
infl_pred_naiv = infl.shift(1, freq='M')
infl_pred_naiv
```

```python
infl_pred_naiv.rename(columns = {'Infl':'infl_pred_naiv'}, inplace=True)
infl_pred_naiv
```

```python
infl_pred = pd.concat([infl,infl_pred_naiv], axis=1)
infl_pred
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
infl_pred_naiv_MSE = mean_squared_error(infl_pred.dropna()['Infl'], infl_pred.dropna()['infl_pred_naiv'])
infl_pred_naiv_MSE
```

```python
infl_pred_naiv_MAE = mean_absolute_error(infl_pred.dropna()['Infl'], infl_pred.dropna()['infl_pred_naiv'])
infl_pred_naiv_MAE
```

```python
infl_pred_naiv_MAPE = mean_absolute_percentage_error(infl_pred.dropna()['Infl'], infl_pred.dropna()['infl_pred_naiv'])
infl_pred_naiv_MAPE
```

# ARIMA

```python
infl2 = infl['1989':'2021']
infl2.describe()
```

```python
infl2.plot(figsize=(50,20),fontsize=40)
```

```python
from statsmodels.tsa import stattools
```

```python
stattools.acf(infl2)
```

```python
stattools.pacf(infl2)
```

```python
from statsmodels.graphics import tsaplots
```

```python
tsaplots.plot_acf(infl2, lags=15, fft=False, zero=False)
```

```python
tsaplots.plot_pacf(infl2, lags=15, zero=False)
```

```python
infl_1diff = infl2.diff(periods=1)
```

```python
stattools.acf(infl_1diff, missing='conservative')
```

```python
stattools.pacf(infl_1diff)
```

```python
infl_1diff.plot(figsize=(50, 20), fontsize=40);
```

```python
tsaplots.plot_acf(infl_1diff, lags=15, fft=False, missing='conservative', zero=False)
```

```python
tsaplots.plot_pacf(infl_1diff, lags=15, method='ld', zero=False)
```
